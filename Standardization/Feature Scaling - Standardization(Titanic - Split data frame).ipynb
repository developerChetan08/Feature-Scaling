```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head(5)
```

# Remove Col:-['PassengerId','Name','Ticket','Cabin']

```python
df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

# Fill NA Value Of Age & Embarked Col:-

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

# Apply OneHotEncoder on Embarked Col:-

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
##Use Sex Col:-
dummies_sex=pd.get_dummies(df['Sex'])
new_sex=dummies_sex.drop(['male'],axis=1).astype(int)
df['Sex']=new_sex
df.head(5)
```

```python
#Use Embarked Col:-
dummies_embarked=pd.get_dummies(df['Embarked'])
dummies_embarked
```

```python
df=pd.concat([df,dummies_embarked],axis=1, join='inner')
```

```python
df.drop(['Embarked'], axis=1,inplace=True)
```

```python
df.head(2)
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Survived', axis=1),
                                                    df['Survived'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

# Use StadardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```
